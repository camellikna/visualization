```python
import matplotlib.pyplot as plt

# Count the number of schools by region
region_counts = df['Region'].value_counts()

# Plot the data
plt.figure(figsize=(10, 6))
region_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Schools by Region', fontsize=16)
plt.xlabel('Region', fontsize=14)
plt.ylabel('Number of Schools', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Count schools by board type
board_type_counts = df['Board Type'].value_counts()

# Create a bar chart
plt.figure(figsize=(8, 6))
board_type_counts.plot(kind='bar', color='coral')
plt.title('Number of Schools by Board Type', fontsize=16)
plt.xlabel('Board Type', fontsize=14)
plt.ylabel('Number of Schools', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Count schools by language
language_counts = df['School Language'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 8))
language_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99'])
plt.title('Distribution of School Languages', fontsize=16)
plt.ylabel('')  # Remove y-label for cleaner look
plt.show()
```

```python
# Count schools by grade range
grade_range_counts = df['Grade Range'].value_counts()

# Create a bar chart
plt.figure(figsize=(10, 6))
grade_range_counts.plot(kind='bar', color='lightgreen')
plt.title('Number of Schools by Grade Range', fontsize=16)
plt.xlabel('Grade Range', fontsize=14)
plt.ylabel('Number of Schools', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Count the number of schools in the top 10 cities
city_counts = df['City'].value_counts().head(10)

# Plot the data
plt.figure(figsize=(10, 6))
city_counts.plot(kind='bar', color='purple')
plt.title('Top 10 Cities with Most Schools', fontsize=16)
plt.xlabel('City', fontsize=14)
plt.ylabel('Number of Schools', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Convert 'Date Open' to datetime
df['Date Open'] = pd.to_datetime(df['Date Open'], errors='coerce')

# Extract the year and count schools by year
df['Year Opened'] = df['Date Open'].dt.year
year_counts = df['Year Opened'].value_counts().sort_index()

# Plot the data
plt.figure(figsize=(10, 6))
year_counts.plot(kind='line', marker='o', color='teal')
plt.title('Number of Schools Opened Over Time', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Number of Schools', fontsize=14)
plt.grid()
plt.tight_layout()
plt.show()
```
