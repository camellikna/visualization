```python
import pandas as pd
import numpy as np
import scipy
import matplotlib.pyplot as plt
```

```python
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0,100,50)

fig , ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y)
```

```python

# Generate bar plot from the output
fig, ax = plt.subplots(figsize=(5, 3))
ax.bar(x,y) #bar plot
```

```python
fig , ax = plt.subplots(figsize=(5,3))
ax.plot(y)
ax.set_title('Toal Growth')
ax.set_ylabel('Total')
ax.set_xlabel('Years')
```

```python
# Generate a histogram plot based on the y
fig, ax = plt.subplots(figsize=(5, 3))
ax.hist(y) # counting the numbers eg. we have 9 from 85
```

```python
font1 = {'family' : 'sans-serif' , 'color' : 'blue' , 'size' : 20}
font2 = {'family' : 'monospace' , 'color' : 'green' , 'size' : 20}

fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y)

ax.set_title('Total',fontdict = font1)
                      
```

```python
# modify our lables and titles' appearance and add fontdict arguments
font1 = {'family':'sans-serif','color':'blue','size':20}
font2 = {'family':'monospace','color':'green','size':14}

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)

ax.set_title('Total growth over time', fontdict = font1)
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
```

```python
# Adjusting dot type
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y,
          marker ="*", # "o"/ "+"
          color = "indigo")
```

```python
# Line style
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y,
          marker ="*", # "o"/ "+"
          color = "indigo",
           linestyle = "--",
           linewidth = 2)
```

```python
# Marker color
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y,
          marker ="*", 
          color = "indigo", # line color
          linestyle = "--",
          linewidth = 2,
           markeredgecolor = "#FA9359", # marker edge color
           markerfacecolor = "#000000") # marker inside color
```

```python
#adding some grid lines for y
fig, ax = plt.subplots(figsize=(5, 3))

ax.plot(x,y,marker='*',
markersize = 12, # NEW ADDITIONS
color = '#7425b9', #code for indigo
linestyle = '--',
linewidth = 2,
markeredgecolor = '#fa9359', # orange color
markerfacecolor = '#000000' ) # black dots inside the stars

ax.grid(axis = 'y') # show the grid for Y
```

```python
# Grid line
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y,
          marker ="*", 
          color = "indigo", 
          linestyle = "--",
          linewidth = 2,
           markeredgecolor = "#FA9359", 
           markerfacecolor = "#000000") 
ax.grid() # both vertical and horizontal grid lines
#ax.grid(axis = "x") # vertical grid lines
#ax.grid(axis = "y") # horizontal grid lines
```

```python
fig , ax = plt.subplots(figsize=(5,3))
ax.plot(x,y,label = "1")

ax.legend(loc = "upper right")
ax.text(10,95,"This is a Table",color='RED')
```

```python
fig , ax = plt.subplots(figsize=(5,3))
ax.plot(x,y,label = "1")

ax.text(1,5,"DATA" , transform = ax.transData)
ax.legend(loc='lower right')
plt.xticks(rotation = 45 , ha='right')
ax.text(10,10, "This Value is Important!" , ha = 'center',color = 'red',size = 5)
```
