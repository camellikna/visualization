```python
#Setting up random seed and sample data
np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0, 100, 50)
y2 = np.random.randint(0, 100, 50)
x2 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y2_bar = np.array([110, 180, 240, 99, 220])
```

```python
# Basic plot
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y1)
ax.plot(x, y2)
```

```python
# Adding labels and legend
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y1, label="Person 1")
ax.plot(x, y2, label="Person 2")
ax.legend(loc="lower right")
```

```python
# Customizing legend with font size and shadow
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y1, label="Person 1")
ax.plot(x, y2, label="Person 2")
ax.legend(loc="lower right", frameon=True, fontsize=12, ncol=2, shadow=True)
```

```python
# Adjusting legend position using bbox_to_anchor
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y1, label="Person 1")
ax.plot(x, y2, label="Person 2")
ax.legend(loc="upper left", bbox_to_anchor=(1, 1))
```

```python
#Text annotation on the plot
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x, y1, label="Person 1")
ax.scatter(x, y2, label="Person 2")
ax.legend(loc="lower right")
ax.text(10, 95, "This value is important!")
```

```python
#Text annotation with custom styling
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x, y1, label="Person 1")
ax.scatter(x, y2, label="Person 2")
ax.legend(loc="lower right")
ax.text(10, 95, "This value is important!", ha="center", color="red", size=20)
```

```python
#Using different coordinate systems for annotations
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.axis([0, 10, 0, 10])
ax.text(1, 5, ". Data: (1, 5)", transform=ax.transData)
ax.text(0.5, 0.1, ". Axes: (0.5, 0.1)", transform=ax.transAxes)
ax.text(0.2, 0.2, ". Figure: (0.2, 0.2)", transform=fig.transFigure)
```

```python
# Adding arrows with annotation
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x, y1, label="Person 1")
ax.scatter(x, y2, label="Person 2")
ax.legend(loc="lower right")
ax.annotate("This is important!", xy=(10, 95), xytext=(20, 94), 
            arrowprops=dict(facecolor="black", shrink=0.05))
```

```python
# Removing ticks and labels
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x, y1, label="Person 1")
ax.scatter(x, y2, label="Person 2")
ax.legend(loc="lower right")
ax.yaxis.set_major_locator(plt.NullLocator())  # Remove ticks
ax.xaxis.set_major_formatter(plt.NullFormatter())  # Remove only labels
```

```python
# Using MaxNLocator for x-axis
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x, y1, label="Person 1")
ax.scatter(x, y2, label="Person 2")
ax.legend(loc="lower right")
ax.xaxis.set_major_locator(plt.MaxNLocator(3))
```

```python
#Customizing tick marks with MultipleLocator
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x, y1, label="Person 1")
ax.scatter(x, y2, label="Person 2")
ax.legend(loc="lower right")
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
```

```python
# Rotating x-ticks and adjusting font style
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x, y1, label="Person 1")
ax.scatter(x, y2, label="Person 2")
ax.legend(loc="lower right")
plt.xticks(rotation=45, ha="right", fontsize=12, font="serif")
```

```python
#Adding an x-axis label with custom font
font1 = {'family': 'serif', 'color': 'indigo'}
plt.xlabel("Shiny New X Axis!", fontsize=18, fontdict=font1)
```

```python
#  Using a custom Matplotlib style
plt.style.use("fivethirtyeight")
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y1)
```

```python
# Creating subplots for scatter and bar charts
fig, (ax1, ax2) = plt.subplots(ncols=2, nrows=1, figsize=(7, 3))
ax1.scatter(x, y1)
ax2.bar(x2, y2_bar)
```

```python
# Adding titles and labels to subplots
font2 = {'family': 'serif', 'size': 14}
ax1.set_title("Important Scatterplot", fontdict=font2)
ax1.set_ylabel("Measure", fontdict=font1)
ax1.set_xlabel("Time", fontdict=font1)
```

```python
# Creating a mosaic of subplots
fig, someaxes = plt.subplot_mosaic([["ax1", "ax3"], ["ax2", "ax3"]], figsize=(7, 4))
someaxes["ax1"].scatter(x, y1)
someaxes["ax2"].bar(x2, y2_bar)
someaxes["ax3"].plot(x, y1)
```

```python
# Adding labels to mosaic subplots
someaxes["ax1"].set_xlabel("A Big Label", fontsize=18)
someaxes["ax2"].set_xlabel("Another Label", fontsize=18)
someaxes["ax3"].set_xlabel("Label 2: 2 Fast 2 Furious", fontsize=18)
```

```python
# Adding labels to each subplot in the mosaic
fig, someaxes = plt.subplot_mosaic([["ax1", "ax3"], ["ax2", "ax3"]], figsize=(7, 4))

# Plotting different types of charts
someaxes["ax1"].scatter(x, y1)
someaxes["ax2"].bar(x2, y2_bar)
someaxes["ax3"].plot(x, y1)

# Adding unique x-axis labels to each subplot
someaxes["ax1"].set_xlabel("A Big Label", fontsize=18)
someaxes["ax2"].set_xlabel("Another Label", fontsize=18)
someaxes["ax3"].set_xlabel("Label 2: 2 Fast 2 Furious", fontsize=18)
```
