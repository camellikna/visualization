```python
tips = sns.load_dataset("tips")
print(tips)
```

```python
sns.lineplot(data=tips, x='total_bill', y='tip')
```

```python
sns.set_style('white')
sns.lineplot(data=tips, x='total_bill', y='tip')
```

```python
tipgraph = sns.lineplot(data=tips, x='total_bill', y='tip')

tipgraph.set(
    title='Tips vs. Total Bill',
    xlabel='Total Bill ($)',
    ylabel='Tip Amount ($)'
)
```

```python
fig = plt.subplots(figsize=(10, 3))

tipgraph = sns.lineplot(
    data=tips,
    x='total_bill',
    y='tip',
    color='hotpink',
    linestyle='-.',
    linewidth=3,
    marker='o',
    markerfacecolor='indigo'
)

tipgraph.set(
    title='Tips vs. Total Bill',
    xlabel='Total Bill ($)',
    ylabel='Tip Amount ($)'
)
```

```python
tipgraph = sns.scatterplot(
    data=tips,
    x='total_bill',
    y='tip',
    style='time',
    hue='day',
    palette=['purple', 'hotpink', 'deepskyblue', 'yellowgreen']
)

tipgraph.set(
    title='Tip vs. Total Bill',
    xlabel='Total Bill ($)',
    ylabel='Tip Amount ($)'
)
```

```python
import plotly.graph_objects as go # 'go' is 'graph objects
x1 = np.array(["Luffy","Zoro","Nami","Usopp","Sanji"])
y1 = np.array([110, 180, 240, 99, 220])

graph = go.Figure()
graph.add_trace(go.Bar(x=x1, y=y1))

```

```python
graph = go.Figure()
graph.add_trace(go.Scatter(x=x1, y=y1, mode='markers'
, # we want points for a scatter plot
marker=dict(
size=15, # point size
color='hotpink'
, # point colour
opacity=1, # point transparency/alpha
line=dict(width=5, color='purple') # point outline
)))
graph.update_layout(
title='Interactive Pirate Plot'
,
xaxis_title='Pirates'
,
yaxis_title='Scores'
,
width=500, height=500)
```

```python
from wordcloud import WordCloud

df = pd.read_csv(
    "https://raw.githubusercontent.com/prasertcbs/basic-dataset/master/movie_quotes.csv",
    on_bad_lines='skip'
)

df
```

```python
text = " ".join(each for each in df.quote)

worldcloud = WorldCloud (background_color = 'white' , colormap = 'inferno').generate (text)

fig,ax = plt.subplot(figsize=(7.3))
ax.imshow(worldcloud ,
           interpolcation = 'bilinear')
ax.axis=("off")
```

```python
text = " ".join(each for each in df.quote)

wordcloud = WordCloud(
    background_color='white',
    colormap='inferno'
).generate(text)

fig, ax = plt.subplots(figsize=(7, 3))
ax.imshow(wordcloud, interpolation='bilinear')
ax.axis("off")
```

```python
from matplotlib_venn import venn2 , venn2_circle , venn2_unweighted

A = set (['apple','banana' , 'watermelon'])
B = set (['pumpkin' , 'blueberry' , 'apple' , 'key lime'])

diagram = venn2 ([A,B] ,
                 set_lbels = ('Fruits' , 'Pie') ,
                 selt_colors = ('blue' , 'red'),
                 alpha = 0.5)

diagram.get_lable_by_id("10").set_text("\n".join(A-B))
diagram.get_lable_by_id("10").set_text("\n".join(A&B))
diagram.get_lable_by_id("10").set_text("\n".join(B-A))
```

```python
from matplotlib_venn import venn2

A = set(['apple', 'banana', 'watermelon'])
B = set(['pumpkin', 'blueberry', 'apple', 'key lime'])

diagram = venn2(
    [A, B],
    set_labels=('Fruits', 'Pies'),
    set_colors=('blue', 'red'),
    alpha=0.5
)

diagram.get_label_by_id("10").set_text("\n".join(A - B))
diagram.get_label_by_id("11").set_text("\n".join(A & B))
diagram.get_label_by_id("01").set_text("\n".join(B - A))
```
